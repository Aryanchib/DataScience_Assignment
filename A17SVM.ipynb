# Support Vector machine

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.simplefilter('ignore')
data=pd.read_csv('/content/sample_data/mushroom.csv')
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
cat_cols=data.select_dtypes(include=['object']).columns
label=LabelEncoder()
```

```python
for col in cat_cols:
    if data[col].dtypes=='object':
        data[col]=label.fit_transform(data[col])
```

```python
data
```

```python
data.drop('Unnamed: 0',axis=1,inplace=True)
data.hist(bins=20,figsize=(30,25))
plt.show()
```

```python
plt.figure(figsize=(30,15))
data.boxplot()
plt.show()
```

```python
corr=data.corr()
plt.figure(figsize=(40,25))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
features=data.drop('class',axis=1)
target=data['class']
```

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=42)
```

```python
sns.pairplot(data)
plt.show()
```

```python
sns.countplot(x='class',data=data)
plt.title('COUNT PLOT')
plt.show()
```

```python
plt.scatter(data['stalk_height'],data['cap_diameter'])
plt.title('SCATTER PLOT')
plt.show()
```

```python
sns.distplot(features)
plt.title('DIST PLOT')
plt.show()
```

```python
from sklearn import svm
from sklearn.svm import SVC
model1=SVC(kernel='linear',gamma='scale',C=5)
```

```python
model1.fit(x_train,y_train)
```

```python
ypred=model1.predict(x_test)
from sklearn.metrics import accuracy_score,f1_score,classification_report,confusion_matrix
print('Accuracy:',accuracy_score(y_test,ypred)*100)
print(classification_report(y_test,ypred))
print('F1 score:',f1_score(y_test,ypred)*100)
```

```python
from sklearn.model_selection import GridSearchCV
params={
    'kernel':['rbf'],
    'gamma':['scale'],
    'C':[0.1,4,5]
}
grid=GridSearchCV(estimator=model1,param_grid=params)
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
con=confusion_matrix(y_test,ypred)
sns.heatmap(con,annot=True)
```
