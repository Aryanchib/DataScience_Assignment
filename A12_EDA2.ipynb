# Discuss the scenarios where each scaling technique is preferred and why.
* Standardization:
Standardization
Definition: Standardization transforms data to have a mean of 0 and a standard deviation of 1.



When to Use:

* Machine Learning Algorithms: Standardization is often used with algorithms that assume normally distributed data, such as linear regression, logistic regression, and support vector machines (SVM).
* Principal Component Analysis (PCA): Standardization is crucial for PCA since it relies on the variance of the data.
* Data with Outliers: Standardization is less sensitive to outliers compared to normalization.
* Normalization (Min-Max Scaling):
Preferred when data needs to be bounded within a specific range (e.g., [0, 1] or [-1, 1]).

When to Use:

* Neural Networks: Normalization is often used with neural networks because it helps in faster convergence during training.
* Distance-Based Algorithms: Algorithms like k-nearest neighbors (KNN) and k-means clustering benefit from normalization as they rely on the distance between data points.
* Data without Outliers: Normalization can be sensitive to outliers, so it's best used when the data does not have extreme values.

Summary:
* Standardization: Use when the algorithm assumes normally distributed data or when dealing with data that has outliers.
* Normalization: Use when the algorithm is distance-based or when working with neural networks.

# 2. Encoding Techniques:
One-Hot Encoding: Convert categorical variables into binary vectors.

```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder
categorical_cols=data.select_dtypes(include='object').columns
categorical_cols
```

```python
for col in categorical_cols:
    b=data[col].unique()
    print(f'{col} had {b}','\n')
```

```python
onehot=OneHotEncoder()
dummies = pd.get_dummies(data[['sex','income']],dtype=int)
dummies
```

```python
label=LabelEncoder()
data['education'] = label.fit_transform(data['education'])
data['workclass'] = label.fit_transform(data['workclass'])
data['marital_status'] = label.fit_transform(data['marital_status'])
data['occupation'] = label.fit_transform(data['occupation'])
data['relationship'] = label.fit_transform(data['relationship'])
data['race'] = label.fit_transform(data['race'])
data['native_country'] = label.fit_transform(data['native_country'])
```

```python
data.drop(data[['sex','income']],axis=1,inplace=True)
data=data.join(dummies,how='right')
data
```

# 4. Feature Selection:

```python
from sklearn.ensemble import IsolationForest
ift=IsolationForest(random_state=10,contamination=0.1)
ift.fit(data)
```

```python
outliers=pd.DataFrame(ift.predict(data))
outliers.columns = ['outliers']
data=data.join(outliers,how='right')
data
```
